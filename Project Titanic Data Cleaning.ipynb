```python
def return_percentage_survived_with_given_Pclass(p_class):
    a=titanic[titanic.Pclass==p_class]
    b=a[titanic.Survived==1]
    return (len(b)/len(a))*100
```

```python
print("Pclass = 1 : ",return_percentage_survived_with_given_Pclass(1))
print("Pclass = 2 : ",return_percentage_survived_with_given_Pclass(2))
print("Pclass = 3 : ",return_percentage_survived_with_given_Pclass(3))
```

```python
#so seems like Pclass=3 has lesser survival rates than Pclass=2 and 1 so important factor hai can't drop
def return_percentage_survived_with_given_gender(gender):
    a=titanic[titanic.Sex==gender]
    b=a[titanic.Survived==1]
    return (len(b)/len(a))*100
```

```python
print("Gender = Male : ",return_percentage_survived_with_given_gender('male'))
print("Gender = Female : ",return_percentage_survived_with_given_gender('female'))
```

```python
def return_percentage_survived_with_given_embarked(embarked):
    a=titanic[titanic.Embarked==embarked]
    b=a[titanic.Survived==1]
    return (len(b)/len(a))*100
```

```python
print("Embarked = S : ",return_percentage_survived_with_given_embarked('S'))
print("Embarked = C : ",return_percentage_survived_with_given_embarked('C'))
print("Embarked = Q : ",return_percentage_survived_with_given_embarked('Q'))
```
